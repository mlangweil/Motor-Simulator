```python
from pynq import Overlay
from pynq import DefaultIP
from pynq import MMIO
import struct

BRAM_ADDR = 0x80010000
ADDRESS_RANGE = 0x1000
ADDRESS_OFFSET = 0x00

class AddDriver(DefaultIP):

    bindto = ['xilinx.com:hls:calculateCoefficients:1.0']

    def __init__(self, description):
        super().__init__(description)

    def process(self, lowerCutoff, upperCutoff, samplingRate):
        self.write(0x10, lowerCutoff)
        self.write(0x18, upperCutoff)
        self.write(0x20, samplingRate)
        self.write(0x28, BRAM_ADDR)



overlay = Overlay('sim_update.bit')   
add_ip = overlay.calculateCoefficients_0


def writeCoefficients(lowerCutoff, upperCutoff, samplingRate):
        overlay.calculateCoefficients_0.process(lowerCutoff, upperCutoff, samplingRate)


def floatToInt(f):
    raw = struct.pack('<f', f)          # float → 4 bytes
    int_val = int.from_bytes(raw, 'little')  # bytes → int
    print("Integer representation:", int_val)
    return int_val


def intToFloat(i):
    raw = i.to_bytes(4, 'little')       # int → 4 bytes
    float_val = struct.unpack('<f', raw)[0]  # bytes → float
    print("Floating-point value:", float_val)
    return float_val








```
